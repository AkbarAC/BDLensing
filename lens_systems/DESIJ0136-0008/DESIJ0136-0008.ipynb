```python
import lenstronomy
import h5py
import numpy as np
import corner
import joblib
import matplotlib.pyplot as plt
from lenstronomy.Workflow.fitting_sequence import FittingSequence
from lenstronomy.Plots.model_plot import ModelPlot
from nbpep8.nbpep8 import pep8
from lenstronomy.Data.coord_transforms import Coordinates
from lenstronomy.Util import mask_util, util

%matplotlib inline
```

```python
with h5py.File('DESIJ0136-0008_F140W.h5', 'r') as f:
    kwargs_data = {}

    for key in f:
        if key == 'exposure_time':
            kwargs_data[key] = f[key][()]
            continue
        kwargs_data[key] = f[key][()]

# print kwargs_data
for key in kwargs_data:
    print(f'{key}: {kwargs_data[key]}')
```

```python
with h5py.File('psf_F140W.h5', 'r') as f:
        
    kwargs_psf = {}
    for key in f:

        kwargs_psf[key] = f[key][()]
kwargs_psf['psf_type'] = 'PIXEL'   
print(kwargs_psf)
```

```python
image_data = kwargs_data["image_data"]
plt.matshow(np.log10(image_data), origin='lower', cmap='cubehelix')
plt.grid(which='both', color='black')
plt.minorticks_on()
plt.grid(which='minor', color='black')


#plt.matshow(np.log10(image_data), origin='lower', cmap='cubehelix');
```

```python
ra_at_xy_0 = kwargs_data["ra_at_xy_0"]
dec_at_xy_0 = kwargs_data["dec_at_xy_0"]
transform_pix2angle = kwargs_data["transform_pix2angle"]

coords = Coordinates(transform_pix2angle, ra_at_xy_0, dec_at_xy_0)
num_pix = len(kwargs_data["image_data"])

x_coords, y_coords = coords.coordinate_grid(num_pix, num_pix)

r = 8  # arcseconds
lens_center_ra = 0
lens_center_dec = 0

mask_outer = mask_util.mask_center_2d(
    lens_center_ra,
    lens_center_dec,
    r,
    util.image2array(x_coords),
    util.image2array(y_coords),
)



# for the galaxy at the center
mask_ext_center = mask_util.mask_ellipse(
    util.image2array(x_coords),
    util.image2array(y_coords),
    lens_center_ra + 0.9, #1.0
    lens_center_dec + .5,
    0.9, #0.8
    1.2, #0.8
    0,
)

# for the galaxy at 11 o clock
mask_ext_eleven = mask_util.mask_ellipse(
    util.image2array(x_coords),
    util.image2array(y_coords),
    lens_center_ra + 1.0,
    lens_center_dec + 1.8,
    0.6,  
    0.4,
    0,
)

mask = (
    (1 - mask_outer)
    * (1 - mask_ext_center)
    * (1 - mask_ext_eleven)
)
mask[mask >= 1] = 1
mask[mask < 0] = 0
mask = mask.reshape(num_pix, num_pix)



# Setting the subplots
fig, axes = plt.subplots(1, 3, figsize=(15, 15))

axes[0].imshow(np.log10(image_data), origin="lower", cmap="cubehelix")
axes[0].grid(True, color="black")
#axes[0].minorticks_on()
#axes[0].grid(which='minor', color='black')
axes[0].set_title("Image")


axes[1].imshow(mask, origin="lower")
axes[1].grid(True, color="black")
axes[1].set_title("Mask")


image_after_mask = np.log10(image_data * mask)

axes[2].imshow(image_after_mask, origin="lower", cmap="cubehelix")
axes[2].grid(True, color="black")
axes[2].set_title("Image after masking")


# Show the plots
plt.show()
```

```python
kwargs_params = {
    'lens_model': lens_params,
    'source_model': source_params,
    'lens_light_model': lens_light_params,
    }

kwargs_constraints = {
    'joint_lens_with_light': joint_lens_with_light,
    'joint_source_with_source': joint_source_with_source,
    }
```

```python
kwargs_likelihood = {'check_bounds': True,
                     'image_likelihood_mask_list': [mask]
                    }

kwargs_numerics = {'supersampling_factor': 1,
                   'supersampling_convolution': False
                   }
```

```python
kwargs_model = {'lens_model_list': lens_model_list,
                'source_light_model_list': source_model_list,
                'lens_light_model_list': lens_light_model_list,
               }

multi_band_list = [[
        kwargs_data, 
        kwargs_psf, 
        kwargs_numerics
        ]]

kwargs_data_joint = {
    'multi_band_list': multi_band_list,
    'multi_band_type': 'single-band'
                    }
```

```python
fitting_seq = FittingSequence(
    kwargs_data_joint, 
    kwargs_model, 
    kwargs_constraints, 
    kwargs_likelihood, 
    kwargs_params
    )

fitting_kwargs_list = [[
    'PSO', 
    {'sigma_scale': 1., 'n_particles': 300, 'n_iterations': 600},
    ]]
 
chain_list = fitting_seq.fit_sequence(fitting_kwargs_list)
kwargs_result = fitting_seq.best_fit()
```

```python
model_plot = ModelPlot(
    multi_band_list, 
    kwargs_model, 
    kwargs_result, 
    arrow_size=0.02,
    cmap_string="cubehelix",
    image_likelihood_mask_list=kwargs_likelihood['image_likelihood_mask_list']
                      )

f, axes = plt.subplots(
    2, 3, figsize=(16, 8), sharex=False, sharey=False
    )

model_plot.data_plot(ax=axes[0,0])
model_plot.model_plot(ax=axes[0,1])
model_plot.normalized_residual_plot(
    ax=axes[0,2], 
    v_min=-3, 
    v_max=3,
    cmap='RdBu_r'
    )

model_plot.source_plot(
    ax=axes[1, 0], 
    deltaPix_source=0.01, 
    numPix=100
    )

model_plot.convergence_plot(
    ax=axes[1, 1], 
    v_max=1, 
    cmap='gist_heat'
    )

model_plot.magnification_plot(
    ax=axes[1, 2], 
    cmap='PiYG'
    )

f.tight_layout()
f.subplots_adjust(
    left=None, 
    bottom=None, 
    unconvolved=True,
    right=None, 
    top=None, 
    wspace=0., 
    hspace=0.05,
    )
plt.show()
```
